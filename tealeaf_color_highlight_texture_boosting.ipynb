```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
def show_image(image):
    cv2_imshow(image)
```

```python

lower_limit = np.array([25,52, 90])
upper_limit = np.array([79, 255, 255])#42

def highlight_green(image, lower_green=lower_limit, upper_green=upper_limit):
    """
    Highlights green color regions in an image while converting other areas to grayscale

    Args:
        image: Input BGR image (numpy array)
        lower_green: Lower HSV threshold for green (default: (35, 40, 40))
        upper_green: Upper HSV threshold for green (default: (85, 255, 255))

    Returns:
        Processed image with green highlighted and grayscale background
    """
    # Convert to HSV color space
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Create mask for green regions
    mask = cv2.inRange(hsv, np.array(lower_green), np.array(upper_green))

    # Convert original image to grayscale (3-channel for blending)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gray_bgr = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

    # Combine color and grayscale regions using mask
    result = np.where(mask[:, :, np.newaxis], image, gray_bgr)

    return result
```

```python
import cv2
import numpy as np

def boost_texture(image, blur_strength=(5, 5), amount=1.5):
    image_float = image.astype(np.float32)

    blurred = cv2.GaussianBlur(image_float, blur_strength, 0)

    mask = image_float - blurred
    enhanced = image_float + amount * mask

    enhanced = np.clip(enhanced, 0, 255).astype(np.uint8)

    return enhanced
```

```python
image = cv2.imread(r'/content/IMG_8857.JPG')
boost_texture_image=boost_texture(image)
final_image=highlight_green(boost_texture_image)
h =np.hstack((image,boost_texture_image,final_image))
show_image(h)
```
